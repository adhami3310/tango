```python
def plot_all(name: str, title = None):
    if title is None:
        title = name.capitalize() + " Access"
    plt.figure(dpi=300)
    plt.scatter(*zip(*tango_data[name]), label='tango')
    plt.scatter(*zip(*multisplay_data[name]), label='multisplay')
    plt.scatter(*zip(*splay_data[name]), label='splay')
    plt.title(title)
    plt.xlabel("Working set size")
    plt.ylabel("Access Cost")
    plt.xscale('log')
    # plt.yscale('log')
    plt.legend()
    plt.show()

plot_all('working_set', 'Working Set Access (n = 131072)')
```

```python
def plot_ratio(name, title=None):
    if title is None:
        title = name.capitalize() + " Access: ratio of tango tree to other algorithms"
    
    tango_x = np.array([x[0] for x in tango_data[name]])
    tango_y = np.array([x[1] for x in tango_data[name]])
    splay_x = np.array([x[0] for x in splay_data[name]])
    splay_y = np.array([x[1] for x in splay_data[name]])
    multisplay_x = np.array([x[0] for x in multisplay_data[name]])
    multisplay_y = np.array([x[1] for x in multisplay_data[name]])

    splay_int_y = np.interp(tango_x, splay_x, splay_y)
    multisplay_int_y = np.interp(tango_x, multisplay_x, multisplay_y)

    # tango_over_splay = [(x[0], x[1]/y[1]) for (x, y) in zip(tango_data[name], splay_data[name]) if x[0] == y[0]]
    # tango_over_multisplay = [(x[0], x[1]/y[1]) for (x, y) in zip(tango_data[name], multisplay_data[name]) if x[0] == y[0]]
    
    plt.figure(dpi=300)
    plt.scatter([],[])
    plt.scatter(tango_x, tango_y / multisplay_int_y, label='multisplay')
    # plt.scatter(tango_x, tango_y / splay_int_y, label='splay')
    plt.title(title)
    plt.xscale('log')
    # plt.yscale('log')
    plt.xlabel("Working set size")
    plt.ylabel("Ratio of tango tree cost to algorithm cost")
    plt.legend()
    plt.show()

plot_ratio('working_set', 'Working Set Access: ratio of tango tree to other algorithms (n = 131072)')
```
