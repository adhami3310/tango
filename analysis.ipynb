```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
font = {'family': 'georgia'}
matplotlib.rc('font', **font)
```

```python
def plot_all(name: str, title = None, passes=25, xfunc="lg"):
    if title is None:
        title = name.capitalize() + " Access"
    plt.figure(dpi=300)
    tango_x, tango_y = map(np.array, zip(*tango_data[name]))
    multisplay_x, multisplay_y = map(np.array, zip(*multisplay_data[name]))
    splay_x, splay_y = map(np.array, zip(*splay_data[name]))

    # y normalization
    tango_y = tango_y / (passes * tango_x)
    multisplay_y = multisplay_y / (passes * multisplay_x)
    splay_y = splay_y / (passes * splay_x)

    # x norm
    if xfunc == "lg":
        tango_x = np.log2(tango_x)
        multisplay_x = np.log2(multisplay_x)
        splay_x = np.log2(splay_x)
    elif xfunc == "lglg":
        tango_x = np.log2(np.log2(tango_x))
        multisplay_x = np.log2(np.log2(multisplay_x))
        splay_x = np.log2(np.log2(splay_x))


    plt.plot(tango_x, tango_y, label='tango', marker='x')
    plt.plot(multisplay_x, multisplay_y, label='multi-splay', marker='^')
    plt.plot(splay_x, splay_y, label='splay', marker='o')
    plt.title(title)
    if xfunc == "lg":
        plt.xlabel("lg n")
    elif xfunc == "lglg":
        plt.xlabel("lg lg n")
    else:
        plt.xlabel("n")
    plt.ylabel("cost / access")
    # plt.xscale('log')
    # plt.yscale('log')
    plt.legend()
    plt.show()

plot_all('sequential', 'Sequential Access (amortized cost)', xfunc="lglg")
```

```python
def plot_ratio(name, title=None, double=False):
    if title is None:
        title = name.capitalize() + " Access: ratio of tango tree to other algorithms"
    
    tango_x = np.array([x[0] for x in tango_data[name]])
    tango_y = np.array([x[1] for x in tango_data[name]])
    splay_x = np.array([x[0] for x in splay_data[name]])
    splay_y = np.array([x[1] for x in splay_data[name]])
    multisplay_x = np.array([x[0] for x in multisplay_data[name]])
    multisplay_y = np.array([x[1] for x in multisplay_data[name]])

    splay_int_y = np.interp(tango_x, splay_x, splay_y)
    multisplay_int_y = np.interp(tango_x, multisplay_x, multisplay_y)

    tango_x = np.log2(np.log2(tango_x))

    # tango_over_splay = [(x[0], x[1]/y[1]) for (x, y) in zip(tango_data[name], splay_data[name]) if x[0] == y[0]]
    # tango_over_multisplay = [(x[0], x[1]/y[1]) for (x, y) in zip(tango_data[name], multisplay_data[name]) if x[0] == y[0]]
        
    if double:
        fig, ax1 = plt.subplots()
        ax1.plot([],[])
        fig.set_dpi(300)
        # plt.plot([],[])
        ln1 = ax1.plot(tango_x, tango_y / multisplay_int_y, label='tango / multi-splay', marker='^')
        ax2=ax1.twinx()
        ax2.plot([],[])
        ax2.plot([],[])
        ln2 = ax2.plot(tango_x, tango_y / splay_int_y, label='tango / splay', marker='o')

        plt.title(title)
        # plt.xscale('log')
        # plt.yscale('log')
        ax1.set_xlabel("lg lg n")
        ax1.set_ylabel("tango cost / multi-splay cost")
        ax2.set_ylabel("tango cost / splay cost")
        
        lns = ln1+ln2
        labs = [l.get_label() for l in lns]
        ax1.legend(lns, labs, loc=0)
        # ax1.legend(loc=0)
        # ax2.legend(loc=0)
        plt.show()
        return
    
    plt.figure(dpi=300)
    plt.plot([],[])
    plt.plot(tango_x, tango_y / multisplay_int_y, label='tango / multi-splay', marker='^')
    plt.plot(tango_x, tango_y / splay_int_y, label='tango / splay', marker='o')
    plt.title(title)
    # plt.xscale('log')
    # plt.yscale('log')
    plt.xlabel("lg lg n")
    plt.ylabel("tango cost / other cost")
    plt.legend()
    plt.show()

plot_ratio('sequential', 'Sequential Access: ratio of tango tree to other algorithms', True)
```
